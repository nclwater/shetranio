```python
from shetranio import Model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the library file for the model
model = Model('../../tests/sample_data/Wansbeck_at_Mitford_Library_File.xml')
```

```python
# Plot discharge at element 45
model.hdf.overland_flow.get_element(45).plot()
```

```python
# Time step to plot
i = 10

# Get the x and y coordinates
x = model.dem.x_coordinates
y = model.dem.y_coordinates
extent = (min(x), max(x), min(y), max(y))

# Clip extra rows and columns added by SHETran
z = model.hdf.ph_depth.values[1:-1, 1:-1, i]

# Set no data values to be NaNs
z[z == -1] = float('NaN')

# Plot the data
plt.imshow(z, extent=extent)
plt.title(model.hdf.ph_depth.times[i])
plt.colorbar(label=model.hdf.ph_depth.long_name)
```
